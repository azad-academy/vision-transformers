```python

import os
assert os.path.exists(f'vit_model.npz')

# Loading and converting the checkpoint.
params = checkpoint.load(f'vit_model.npz')
params['pre_logits'] = {}  # Restoring the empty leaf for Flax.
```

```python
import wget
# Downloading imagenet labels.
wget.download("https://storage.googleapis.com/bit_models/ilsvrc2012_wordnet_lemmas.txt")
imagenet_labels = dict(enumerate(open('ilsvrc2012_wordnet_lemmas.txt')))
```

```python
model_config = models_config.MODEL_CONFIGS["ViT-B_32"]

model = models.VisionTransformer(num_classes=1000, **model_config)
```

### Downloading Test Images

```python
# Downloading few random images for testing.

def download_sample_images(num_imgs):
  for i in range(0,num_imgs):
    filename = f"{i}.jpg"
    if os.path.exists(filename):
      os.remove(filename)
    wget.download("https://picsum.photos/384",out=filename)

num_imgs = 3
download_sample_images(num_imgs)
```

```python
from PIL import Image
from matplotlib import pyplot as plt

# Load images from disk and display them
def load_images(img_dir,num_imgs):
  imgs = []
  fig,axes = plt.subplots(1,num_imgs,figsize=(20,10))
  for i in range(0,num_imgs):
    img = Image.open(img_dir+f'{i}.jpg')
    imgs.append(img)
    axes[i].imshow(img)
    axes[i].set_axis_off()
  fig.tight_layout()
  return imgs

imgs = load_images("./",num_imgs)
```

### Predicting Label

```python
# Predicting the labels of the images

img_labels = []
img_probs = []
for i in range(0,num_imgs):
  logits, = model.apply(dict(params=params), (np.array(imgs[i]) / 128 - 1)[None, ...], train=False)
  probs = np.array(jax.nn.softmax(logits))
  pred_indices = probs.argsort()[:-6:-1]  #taking the top 5 with highest probability
  labels = [imagenet_labels[i] for i in pred_indices]
  p = probs[pred_indices]
  img_labels.append(labels)
  img_probs.append(p)
```

```python
#import importlib
from plt_tools import plot_predictions
#importlib.reload(plot_predictions)

plot_predictions(img_labels,img_probs,1,3,1200,350,True)
del model # Clearing GPU memory
```

## 2. Image Caption Prediction

In this demo, we shall use a pre-trained transformer network on COCO dataset. The transformer network predicts the caption of a given image.

```python
import torch
cap_model = torch.hub.load('saahiluppal/catr', 'v3', pretrained=True)  
```

```python
from datasets import coco, utils
from transformers import BertTokenizer
from utils import *

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')    
img_captions = []

for i in range(0,num_imgs):

  start_token = tokenizer.convert_tokens_to_ids(tokenizer._cls_token)
  end_token = tokenizer.convert_tokens_to_ids(tokenizer._sep_token)

  #creating an initial caption in embedded format
  #i.e. All zero vector with start_token as the first element
  init_caption, cap_mask = create_caption_and_mask(start_token, config.max_position_embeddings)      

  image = coco.val_transform(imgs[i])
  image = image.unsqueeze(0)

  output = evaluate(cap_model,image,init_caption,cap_mask)
  result = tokenizer.decode(output[0].tolist(), skip_special_tokens=True)
  caption = result.capitalize()

  img_captions.append(caption)
  print(caption)
```

```python
#Displaying the captions with the images
fig,axes = plt.subplots(1,num_imgs,figsize=(20,10))
for i in range(0,num_imgs):
  axes[i].imshow(imgs[i])
  axes[i].set_axis_off()
  axes[i].set_title(img_captions[i],fontsize=15,fontweight="bold")
fig.tight_layout()
```

## 3. Contrastive Language-Image Pre-training

```python
# Downloading & installing the CLIP model
!pip install git+https://github.com/openai/CLIP.git
```

```python
torch.cuda.empty_cache()
#Loading the CLIP model
import clip
model, preprocess = clip.load("ViT-B/32")
model.cuda().eval()
input_resolution = model.visual.input_resolution
context_length = model.context_length
vocab_size = model.vocab_size

print("Model parameters:", f"{np.sum([int(np.prod(p.shape)) for p in model.parameters()]):,}")
print("Input resolution:", input_resolution)
print("Context length:", context_length)
print("Vocab size:", vocab_size)
```

```python
#Loading sample images
num_imgs = 5
imgs = load_images("./images/",num_imgs)
```

```python
#Writing descriptions for each sample image in the images directory
descriptions = {
    0: "An image of an office desk. There are three chairs and a table. An LCD screen and a Laptop computer is on the desk. There are also a notebook, a headphone, a keyboard, a glass, two plants and a lamp on the table.",
    1: "An image of a baseball game. One player is catching the ball and another player is running towards the base.",
    2: "An image of a kitchen counter. There are two knives, a plate, three potatoes, a coffee machine, a coffee grinder and a toaster on the counter.",
    3: "Two women are sitting across the table and talking to each other. There is a laptop, a glass, and phone on the table. One woman is looking on the laptop while another woman is looking at the first woman.",
    4: "An asian woman with glasses is sitting in the kitchen and eating a sandwitch. There is a plate and a cup on the table.",    
}
```

```python
processed_images = []

from textwrap import wrap
fig,axes = plt.subplots(1, num_imgs, figsize=(16, 5))

for i in range(0,5):
  processed_images.append(preprocess(imgs[i]))
  axes[i].imshow(imgs[i])
  axes[i].set_title("\n".join(wrap(f"{descriptions[i]}",30)))
  axes[i].set_axis_off()
fig.tight_layout()
fig.subplots_adjust(top=0.8)
```

```python
#Tokenizing the text and images
texts = descriptions.values()
image_inputs = torch.tensor(np.stack(processed_images)).cuda()
text_tokens = clip.tokenize(["" + desc for desc in texts]).cuda()

#Computing the image and text features by running a forward pass on the model

with torch.no_grad():
    img_features = model.encode_image(image_inputs).float()
    txt_features = model.encode_text(text_tokens).float()

```

```python
#Computing the pairwise Cosine Similarity between the text snippets and the images

img_features /= img_features.norm(dim=-1, keepdim=True)
txt_features /= txt_features.norm(dim=-1, keepdim=True)
similarity = txt_features.cpu().numpy() @ img_features.cpu().numpy().T
```

```python
#Plotting the Cosine Similarity Matrix

plt.figure(figsize=(20, 14))
plt.imshow(similarity, vmin=0.1, vmax=0.3,cmap='gray')
ylabels = ["\n".join(wrap(txt,50)) for txt in texts]
plt.yticks(range(num_imgs), ylabels, fontsize=18)
plt.xticks([])
for i, image in enumerate(imgs):
    plt.imshow(image, extent=(i - 0.5, i + 0.5, -1.6, -0.6), origin="lower")
for x in range(similarity.shape[1]):
    for y in range(similarity.shape[0]):
        plt.text(x, y, f"{similarity[y, x]:.2f}", ha="center", va="center", size=12)

for side in ["left", "top", "right", "bottom"]:
  plt.gca().spines[side].set_visible(False)

plt.xlim([-0.5, num_imgs - 0.5])
plt.ylim([num_imgs + 0.5, -2])

plt.title("Similarity Between Text Snippet and Image", size=20)
```
